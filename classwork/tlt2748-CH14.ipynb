# Challenge 14

Finite difference methods

## Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Functions

```python
a, b = (0, 12)
n = 100
h = (b-a) / (n-1)

f = lambda x: np.cos(x)
central = lambda x_arr: [(f(x+h)-f(x-h)) / (2*h) for x in x_arr]
forward = lambda x_arr: [(f(x+h)-f(x)) / h for x in x_arr]
backward = lambda x_arr: [(f(x)-f(x-h)) / h for x in x_arr]
```

## Graphing

True derivative vs 3 methods

```python
x = np.array([h*i for i in range(n)])
plt.figure(figsize=(12,8))
plt.plot(x, -np.sin(x), 'b-', label="True Derivative")
plt.plot(x, central(x), 'r-', label="Central Method")
plt.plot(x, forward(x), 'g-', label="Forward Method")
plt.plot(x, backward(x), 'y-', label="Backward Method")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Cosine Derivatives")
```
