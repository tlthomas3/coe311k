# Challenge 1

Create 1 plot with 4 different subplots.

```python
import matplotlib.pyplot as plt
import numpy as np
from random import random
```

```python
plt.subplot(2,2,1)
plt.plot([1,4], [4,1], 'r-')
plt.subplot(2,2,2)
plt.scatter((x := range(100)), [np.random.rand() for _ in range(len(x))])
plt.subplot(2,2,3)
plt.bar(range(3), [0.5,1,2], color='b')
plt.subplot(2,2,4)
plt.stem([1,2,3,4,5], [6,7,8,3,5], 'k-', use_line_collection=True)
plt.show()
```
