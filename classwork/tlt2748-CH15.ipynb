Euler's method step size vs error

## Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Calculating and graphing

```python
plt.figure(figsize=(12,6))
xs, ys = ([], [])
for h in np.arange(0.01, 1.0, 0.05):
    x_values = np.arange(0, 2, h)
    y_values = np.zeros_like(x_values)
    y_values[0] = 1

    for i in range(len(x_values)-1):
        y_values[i + 1] = (1 - 2*h) * y_values[i]
    xs.append(x_values)
    ys.append(y_values)
    plt.plot(x_values, y_values, label=f"h={h:.2f}")

plt.xlabel('x')
plt.ylabel('y')
plt.title("Approximation of dy/dx = -2y using Euler's Method")
plt.legend()
plt.show()
```

## Errors

```python
print("h=0.01: Truth")
for x_values, y_values in zip(xs[1:], ys[1:]):
    h = x_values[1]-x_values[0]
    true_y = lambda y: min(ys[0], key=lambda x:abs(x-y))
    errors = [float(y-true_y(y)) for y in y_values]
    print(f"\nh={h:.2f}:")
    print(f"Global Error: {sum(errors):.6f}")
    print("Local Errors:")
    plt.figure(figsize=(3,1))
    plt.plot(x_values, errors, 'bo-')
    plt.xlim(0,2)
    plt.xlabel("x")
    plt.ylabel("Error")
    plt.title(f"h={h:.2f} Error")
    plt.show()
```
