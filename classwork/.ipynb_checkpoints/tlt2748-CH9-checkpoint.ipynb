# Challenge 9

Least squares

## Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Part 1

```python
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m:.2f}, Intercept: {c:.2f}")

xs = np.array([X[0],X[-1]])
plt.scatter(X, Y)
plt.plot(xs, m*xs + c, 'r-')
plt.show()
```

## Part 2

Randomize the order of the input vectors

```python
np.random.seed(0)
X, Y = np.random.random(size=(2,5))*5
A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m:.2f}, Intercept: {c:.2f}")

xs = np.array([min(X),max(X)])
plt.scatter(X, Y)
plt.plot(xs, m*xs + c, 'r-')
plt.show()
```

## Part 3

```python
m = 3
b = -2
x = np.arange(0, 1, 0.01)
y = m*x + b + (2*np.random.random(size=len(x)) - 1)
A = np.vstack([x, np.ones(len(x))]).T
s, r, _, _ = np.linalg.lstsq(A, y, rcond=None)

print(f"Original     : y = {m:.2f}x + {b:.2f}")
print(f"Least Squares: y = {s[0]:.2f}x + {s[1]:.2f}")

xs = np.array([min(x),max(x)])
plt.scatter(x, y)
plt.plot(xs, s[0]*xs + s[1], 'r-')
plt.show()
```
