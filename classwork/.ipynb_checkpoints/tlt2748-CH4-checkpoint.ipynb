# Challenge 4

Code the Leibniz Series for 𝜋 in Python.

Run through the first 5,000 terms of the series.

Every 100 terms, print out the relative error between the current running calculated value of 𝜋 and numpy.pi

As each term is added to the summation, what is happening to our error?

## Setup

```python
import numpy as np
```

## Leibniz Series Error

```python
my_quarter_pi: float = 0.0
for n in range(5000):
    my_quarter_pi += (1-2*(n%2))/(2*n + 1)
    if not n%100:
        print(abs((4.0 * my_quarter_pi) - np.pi) / np.pi)
```

## Discussion

Error decreases as we add more terms to the series.
