# Challenge 6

Solve 2 sets of equations with Jacobi and Gauss-Seidel and compare results.

## Setup

```python
import numpy as np
from time import perf_counter

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

## Testing

```python
tolerance = 0.000001 # 1e-6

A = {1: np.array([
        [10, -1, 2, 0, 0],
        [-1, 11, -1, 3, 0],
        [2, -1, 10, -1, 0],
        [0, 3, -1, 8, -2],
        [0, 0, 0, -2, 9]
        ]),
     2: np.array([
        [1, 2, 3, 0, 0],
        [2, 1, 2, 3, 0],
        [3, 2, 1, 2, 3],
        [0, 3, 2, 1, 2],
        [0, 0, 3, 2, 1]
        ])}

b = {1: np.array([14, 30, 26, 25, 37]),
     2: np.array([14, 22, 33, 26, 22])}

for i in [1, 2]:
    t_start = perf_counter()
    _, n1 = jacobi(A[i], b[i], np.array([0, 0, 0, 0, 0]), tolerance, 1000)
    t_middle = perf_counter()
    _, n2 = gauss_seidel(A[i], b[i], np.array([0, 0, 0, 0, 0]), tolerance, 1000)
    t_end = perf_counter()
    
    print(f"Case {i}:")
    print(f"      Jacobi: {n1}{' (max)' if (n1==1000) else ''} iterations, {t_middle-t_start:.6f} seconds")
    print(f"Gauss-Seidel: {n2}{' (max)' if (n2==1000) else ''} iterations, {t_end - t_middle:.6f} seconds")
```

## Discussion

Case 1 works great for both converging methods since it is diagonally dominant. Jacobi needed 1 less iteration but still took slightly longer to run than Gauss-Seidel.

Case 2 is not diagonally dominant and therefore does not work with either converging method, which is why both hit max iterations and multiple runtime warnings popped up.
