2nd-order Euler

## Demo Code

```python
import numpy as np
import matplotlib.pyplot as plt

k = 10.0
h = 0.01
x_start = 0
x_end = 10
y_0 = 1.0
v_0 = 0.0

x_values = np.arange(x_start, x_end + h, h)
y_values = np.zeros(len(x_values))
v_values = np.zeros(len(x_values))

y_values[0] = y_0
v_values[0] = v_0

for i in range(1, len(x_values)):
    x_n = x_values[i - 1]
    y_n = y_values[i - 1]
    v_n = v_values[i - 1]
    y_values[i] = y_n + h * v_n
    v_values[i] = v_n + h * (-k * y_n)

plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label='Position y(x)')
plt.plot(x_values, v_values, label='Velocity v(x)',
linestyle='--')
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.legend()
plt.grid(True)
plt.show()
```

- Play with the value of K, what are you seeing?
K and the amplitude of velocity have a positive correlation while the amplitude of position stays near 1. Also, K and the frequency of oscillation have a positive correlation. This makes sense because that is how a spring should work when changing the spring constant K.
