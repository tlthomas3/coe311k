2nd-order Euler

## Demo Code

```python
import numpy as np
import matplotlib.pyplot as plt

# Otherwise the plot doesn't show up sometimes for me
%matplotlib inline

k = 1.0
x_start = 0
x_end = 10
y_0 = 1.0
v_0 = 0.0

plt.figure(figsize=(10, 5))
for h in [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]:
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    y_values[0] = y_0
    v_values[0] = v_0

    for i in range(1, len(x_values)):
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)
    plt.plot(x_values, y_values, label=f"h={h:.2f}")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.legend()
plt.grid(True)
plt.show()
```
