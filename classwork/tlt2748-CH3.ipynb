## Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Monte Carlo Algorithm

```python
def monte_carlo(n: int) -> float:
    # Count the number of random points that land inside the unit circle
    in_circle: int = 0
    for _ in range(n):
        # Check if a random point in first quadrant is in the unit circle
        x, y = np.random.random(2)
        if np.sqrt(np.power(x,2) + np.power(y,2)) <= 1.0:
            in_circle += 1
    # Only one quadrant used so must multiply by 4
    return 4.0 * in_circle / n
```

## Error Determination

```python
steps = np.array([10]+list(range(20,10000,20)))
values = np.zeros_like(steps, dtype=float)
for i in range(len(steps)):
    # Calculate the absolute error between monte_carlo(n) and np.pi
    values[i] = monte_carlo(steps[i])
error_abs = abs(values - np.pi)
error_rel = abs(values - np.pi) / np.pi
```

## Graphing

```python
plt.title("Monte Carlo Error")
plt.subplot(2,1,1)
plt.plot(steps, error_abs, 'b-')
plt.ylabel("Absolute")
plt.xlabel("Steps")
plt.subplot(2,1,2)
plt.plot(steps, error_rel, 'b-')
plt.ylabel("Relative Error")
plt.xlabel("Steps")
plt.show()
```
