## Setup

```python
import numpy as np
import matplotlib.pyplot as plt
# Needed for sine's taylor series
factorial = lambda x: 1 if x<=1 else x*factorial(x-1)
```

## Taylor Series Function

```python
def taylor_series_sine(x: float, terms: int) -> float:
    my_sine: float = 0.0
    for n in range(terms):
        my_sine += (1-2*(n%2)) * np.power(x,2*n + 1) / factorial(2*n + 1)
    return my_sine
```

## Computation

```python
x_vals = np.arange(0,10,0.1)
my_sines = np.zeros((len(x_vals),4))
for j in range(4):
    for i in range(len(x_vals)):
        my_sines[i,j] = taylor_series_sine(x_vals[i],j+1)
real_sines = np.sin(x_vals)
```

## Error Determination

```python
plt.figure(figsize=(12,12))
plt.suptitle("Absolute Error in Sine Taylor Series Approximation", fontsize=16)
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.plot(x_vals, np.abs(real_sines - my_sines[:,i]), 'b-')
    plt.xlabel("x")
    plt.ylabel(f"Absolute Error ({i+1} Taylor Series terms)", fontsize=14)
plt.show()
```

## Discussion

As we include more terms, there are two things that change. First, the error near 0 decreases, i.e. the estimated sin(x) approaches the real value of sin(x). Second, the further away from 0 we go, the error increases as we add more terms, i.e. the error of the 4-term approximation is larger than the error of the 2-term approximation at x=10.

The error does not decrease uniformly - in fact, the error only decreases for small values of x, while it increases for large values of x.
